2.设计模型（nn.Module）    
3.使用PyTorch API构造loss和optimizer  
4.训练(forward, backward, update)  

```python
#准备数据集
import torch

x_data = torch.Tensor([[1.0], [2.0], [3.0]])
y_data = torch.Tensor([[2.0], [4.0], [6.0]])
```

```python
#设计模型
#torch.nn.Linear(in_features, out_features, bias=True)
#in_features：输入维度
#out_features：输出维度
class LinearModel(torch.nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.linear = torch.nn.Linear(1, 1) #包含了权重和偏置
    
    def forward(self, x):
        y_pred = self.linear(x) #y=w * x + b
        return y_pred

model = LinearModel()
```

```python
#构造损失函数和优化器
#torch.nn.MSELoss(size_average=True, reduce=True)
#size_average：是否求均值
#reduce：True->返回标量，False->返回向量
#torch.optim.SGD(params, lr=<required parameter>, momentum=0, dampening=0, weight_decay=0, nesterov=False)
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
#训练
#1.算y_hat
#2.算loss
#3.backward
#4.更新
for epoch in range(1000):
    y_hat = model(x_data)
    loss = criterion(y_hat, y_data)
    print(epoch, loss)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
print('w = ', model.linear.weight.item())
print('b = ', model.linear.bias.item())
```

```python
x_test = torch.Tensor([[4.0]])
y_test = model(x_test)
print('y_pred=', y_test.data)
```
