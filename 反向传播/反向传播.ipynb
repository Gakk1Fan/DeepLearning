使用Pytorch实现一个线性模型  
y_hat = w * x

```python
import torch
#数据集
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
#设置初试w
w = torch.Tensor([1.0])
#w是需要计算梯度的，默认为False
w.requires_grad = True
```

```python
def forward(x):
    return w * x
```

```python
def loss(x, y):
    y_hat = forward(x)
    return (y - y_hat) ** 2
```

```python
print('训练开始前:',4, forward(4).item())
for epoch in range(200):
    for x, y in zip(x_data, y_data):
        l = loss(x, y)
        l.backward()
        #print('\t', x, y, w.grad.item())
        w.data = w.data - 0.02 * w.grad.data
        w.grad.data.zero_()
    print('epoch=', epoch, 'loss=', l.item())
print('训练后:',4, forward(4).item())
```
